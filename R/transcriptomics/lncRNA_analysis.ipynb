```python
import pandas as pd
```

```python
lncRNA_list_experimental_df = pd.read_excel('../../data/LNCRNAs/LONGLONG_List.xlsx')
res_v18_df = pd.read_csv('../../data/RNA-seq_data_chip/res_v18.csv', index_col=1).drop('Unnamed: 0', axis=1)
res_v48_df = pd.read_csv('../../data/RNA-seq_data_chip/res_v48.csv', index_col=1).drop('Unnamed: 0', axis=1)
gene_annotations = pd.read_csv('gene_annotation.csv', index_col=1,error_bad_lines=False)
```

```python
res_v18_FC = []
res_v18_padj = []
res_v48_FC = []
res_v48_padj = []

for lncRNA in list(lncRNA_list_experimental_df['symbol']):
    try:
        res_v18_FC.append(res_v18_df.loc[lncRNA,'log2FoldChange'])
        res_v18_padj.append(res_v18_df.loc[lncRNA,'padj'])
        res_v48_FC.append(res_v48_df.loc[lncRNA,'log2FoldChange'])
        res_v48_padj.append(res_v48_df.loc[lncRNA,'padj'])
    except KeyError: 
        res_v18_FC.append('')
        res_v18_padj.append('')
        res_v48_FC.append('')
        res_v48_padj.append('')
```

```python
lncRNA_RNAseq_df = lncRNA_list_experimental_df.copy()
lncRNA_RNAseq_df['log2 fold change (18hrs)'] = res_v18_FC
lncRNA_RNAseq_df['adjusted p-value (18hrs)'] = res_v18_padj
lncRNA_RNAseq_df['log2 fold change (48hrs)'] = res_v48_FC
lncRNA_RNAseq_df['adjusted p-value (48hrs)'] = res_v48_padj
```

```python
# lncRNA_RNAseq_df.to_csv('../../data/lncRNAs/RNAseq_comparison.csv')
annotation = []

for lncRNA in lncRNA_RNAseq_df['symbol']:
    try:
        annotation.append(gene_annotations.loc[lncRNA,'GENENAME'])
    except KeyError: 
        annotation.append('')
lncRNA_RNAseq_df['annotation'] = annotation
```

```python
lncRNA_RNAseq_df.to_csv("../../data/lncRNAs/lncRNA_RNAseq.csv")
```
