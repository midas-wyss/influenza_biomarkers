```python
import pandas as pd
import numpy as np
```

```python
COPD_metadata = pd.read_excel("../../../data/COPD_external_RNAseq/COPD_metadata_v2.xlsx")
COPD_metadata
```

```python
# initializing lists that contain metadata we care about
sample_name = []
biopsy_location = []
disease_stat = []
disease_location = []
COPD_counter = 0 # case
control_counter = 0 # cont

# dictionary that will simplify numbering samples
counter_dict = {"copd_MO":0,
                "copd_BE":0,
                "copd_WB":0,
                "control_MO":0,
                "control_BE":0,
                "control_WB":0}


for sample_name_orig in COPD_metadata.columns:
    
    # extract disease status (COPD vs control)
    if COPD_metadata.loc[8, sample_name_orig].split(':')[1][1:] == "case":
        group = "copd"
    else:
        group = "control"
    disease_stat.append(group)
    
    # extract location from which biopsy was taken
    if COPD_metadata.loc[1, sample_name_orig].split(':')[1][1:] == "alveolar macrophage":   
        bio_loc = "MO"
    elif COPD_metadata.loc[1, sample_name_orig].split(':')[1][1:] == "bronchial epithelium":   
        bio_loc = "BE"
    else: 
        bio_loc = "WB"
    biopsy_location.append(bio_loc)
    
    # update dictionary 
    sample_name_group = group + "_" + bio_loc
    counter_dict[sample_name_group] = counter_dict[sample_name_group] + 1

    # updating sample name and disease location lists
    sample_name.append("_".join([group, bio_loc, str(counter_dict[sample_name_group])]))
    disease_location.append(group+"_"+bio_loc)
COPD_metadata_dict = {"sample_name_original": list(COPD_metadata.columns),
                     "GSM_ID": list(COPD_metadata.loc[0,]),
                     "sample_name": sample_name,
                     "biopsy_location": biopsy_location,
                      "group": disease_stat,
                      "group_location": disease_location
                     }

COPD_metadata_df = pd.DataFrame(COPD_metadata_dict)
```

```python
COPD_metadata_df.to_csv("../../../data/COPD_external_RNAseq/COPD_metadata_external_final.csv")
```
