## 47-plex data analysis

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
import statsmodels
import statsmodels.stats.multitest as multi
from scipy import stats
```

```python
# loading in required data sets (gene ids, count matrix, metadata, DESeq2 output)

# loading in gene ids remaining after filtering data
with open('../../../data/gene_lists/gene_ids_chip_preprocessed.txt', 'r') as gene_ids:
    gene_ids = gene_ids.readlines()
    gene_ids_chip = [s.strip('\n') for s in gene_ids]

# loading count matrix and setting gene names as the index
count_data_chip_df = pd.read_csv('../../../data/RNA-seq_data_chip/original_chip_data/count_data_organ_on_chip_preprocessed.csv')
index_conversion_dict = dict(zip(list(count_data_chip_df.index), gene_ids_chip))
count_data_chip_df = count_data_chip_df.rename(index=index_conversion_dict)

# loading sample metadata
sample_metadata_chip_df = pd.read_csv('../../../data/RNA-seq_data_chip/original_chip_data/organ_on_chip_metadata.csv')

# getting column names for virus and control groups
sample_to_group = dict(zip(sample_metadata_chip_df.sample_name, 
                                   sample_metadata_chip_df.group_time))
samples_control18_virus_18 = [key for key, value in sample_to_group.items() if 'control_18' in value.lower() or 'virus_18' in value.lower()]
samples_control48_virus_48 = [key for key, value in sample_to_group.items() if 'control_48' in value.lower() or 'virus_48' in value.lower()]

# differential expression matrix
DE_chip_c18_v18_df = pd.read_csv('../../../data/RNA-seq_data_chip/original_chip_data/DE_comparisons/res_v18.csv', index_col=1).iloc[:,1:]
DE_chip_c48_v48_df = pd.read_csv('../../../data/RNA-seq_data_chip/original_chip_data/DE_comparisons/res_v48.csv',index_col=1).iloc[:,1:]

# cytokine data from 47-plex
cytokine_data_df = pd.read_excel('../../../data/cytokine_data/47_plex_all.xlsx', index_col=0)
```

```python
# creating figures of multi-plex and RNA-seq data
MHT = True
genes_filtered_out = []
# following order will be used throughout: control_18, virus_18, control_48, virus_48
conditions_1 = ['control_18hrs', 'virus_18hrs', 'control_48_hrs', 'virus_48hrs']
conditions_2 = ['18hrs', '48hrs']
empty = ['', '','', '']

protein_18hr_raw_pvals = []
protein_48hr_raw_pvals = []

# statistical testing
for cytokine in list(cytokine_data_df.columns):
    name = cytokine.split('(')[0].replace('/','_')
    # cytokine abundance data
    cytokine_control_18 = cytokine_data_df.loc[:,cytokine][1:4]
    cytokine_virus_18 = cytokine_data_df.loc[:,cytokine][4:7]
    cytokine_control_48 = cytokine_data_df.loc[:,cytokine][7:10]
    cytokine_virus_48 = cytokine_data_df.loc[:,cytokine][10:13]
    
    protein_18hr_raw_pvals.append(stats.ttest_rel(cytokine_virus_18, cytokine_control_18)[1])
    protein_48hr_raw_pvals.append(stats.ttest_rel(cytokine_virus_48, cytokine_control_48)[1])
    
protein_18hr_raw_noNan_pvals=[x if x <=1 and x >= 0 else 1 for x in protein_18hr_raw_pvals]
protein_48hr_raw_noNan_pvals=[x if x <=1 and x >= 0 else 1 for x in protein_48hr_raw_pvals]

if MHT:
    _,protein_18hr_adj_pvals,_,_ = multi.multipletests(protein_18hr_raw_noNan_pvals, alpha = 0.05, method = "fdr_bh" )
    _,protein_48hr_adj_pvals,_,_ = multi.multipletests(protein_48hr_raw_noNan_pvals, alpha = 0.05, method = "fdr_bh" )
else:
    protein_18hr_adj_pvals = protein_18hr_raw_pvals
    protein_48hr_adj_pvals = protein_48hr_raw_pvals

protein_18hr_adj_pvals_dict = dict(zip(list(cytokine_data_df.columns), list(protein_18hr_adj_pvals)))
protein_48hr_adj_pvals_dict = dict(zip(list(cytokine_data_df.columns), list(protein_48hr_adj_pvals)))

def log2_division(n, d):
    if n == 0:
        return(0)
    elif d == 0:
        return 0
    else:
        return(np.log2(n/d))
    
def label_diff(i,j,text,X,Y, axis_name, err):
    x = (X[i]+X[j])/2
    y = 1.1*max(Y[i]+0.75*err[i], Y[j]+0.75*err[j])
    dx = abs(X[i]-X[j])

    props = {'connectionstyle':'bar','arrowstyle':'-',\
                 'shrinkA':10,'shrinkB':10,'linewidth':1}
    axis_name.annotate(text, xy= (X[i],1.3*(Y[i]+0.5*err[i])), zorder=10)
    axis_name.annotate('', xy=(X[i],y), xytext=(X[j],y), arrowprops=props)
    
def p_val_message(p_val):
    if p_val >0.05:
            return("n.s.")
    elif 0.01 < p_val and p_val <=0.05:
        return("*")
    else:
        return("**")



colors = ['blue', 'red', 'darkcyan', 'hotpink']
for cytokine in list(cytokine_data_df.columns):
    name = cytokine.split('(')[0].replace('/','_')
    try:
        # cytokine abundance data
        cytokine_control_18 = cytokine_data_df.loc[:,cytokine][1:4]
        cytokine_virus_18 = cytokine_data_df.loc[:,cytokine][4:7]
        cytokine_control_48 = cytokine_data_df.loc[:,cytokine][7:10]
        cytokine_virus_48 = cytokine_data_df.loc[:,cytokine][10:13]

        # cytokine summary data
        cytokine_means = [np.mean(cytokine_control_18), np.mean(cytokine_virus_18), np.mean(cytokine_control_48), np.mean(cytokine_virus_48)]
        cytokine_std = [np.std(cytokine_control_18), np.std(cytokine_virus_18), np.std(cytokine_control_48), np.std(cytokine_virus_48)]
        
        # cytokine log2 fold change
        cytokine_fold_change = [log2_division(cytokine_means[1], cytokine_means[0]),log2_division(cytokine_means[3], cytokine_means[2])]
        
        # finding the corresponding gene name for current cytokine
        prot_idx = cytokine_data_df.columns.get_loc(cytokine)
        gene_name = cytokine_data_df.loc['gene_name',:][prot_idx]

        control_18_names = [samp for samp in samples_control18_virus_18 if 'C' in samp]
        control_48_names = [samp for samp in samples_control48_virus_48 if 'C' in samp]

        virus_18_names = [samp for samp in samples_control18_virus_18 if 'V' in samp]
        virus_48_names = [samp for samp in samples_control48_virus_48 if 'V' in samp]

        # transcript abundance data
        RNA_control_18 = count_data_chip_df[control_18_names].loc[gene_name]
        RNA_virus_18 = count_data_chip_df[virus_18_names].loc[gene_name]
        RNA_control_48 = count_data_chip_df[control_48_names].loc[gene_name]
        RNA_virus_48 = count_data_chip_df[virus_48_names].loc[gene_name]

        # transcript summary data
        RNA_means = [np.mean(RNA_control_18), np.mean(RNA_virus_18), np.mean(RNA_control_48), np.mean(RNA_virus_48)]
        RNA_std = [np.std(RNA_control_18), np.std(RNA_virus_18), np.std(RNA_control_48), np.std(RNA_virus_48)]

        # transcript fold-changes from differential expression analysis
        RNA_18hr_fold_change = DE_chip_c18_v18_df.loc[gene_name,'log2FoldChange']
        RNA_48hr_fold_change = DE_chip_c48_v48_df.loc[gene_name,'log2FoldChange']
        RNA_fold_changes = [RNA_18hr_fold_change, RNA_48hr_fold_change]

        # plotting figures
        figure(figsize=(8, 8), dpi = 300)  
        
        # cytokine abundance level subplot
        ax_cyto_abund = plt.subplot(2, 2, 1)
        ax_cyto_abund.spines["top"].set_visible(False)  
        ax_cyto_abund.spines["right"].set_visible(False) 
        y_pos = np.arange(len(conditions_1))
        ax_cyto_abund.get_xaxis().tick_bottom()  
        ax_cyto_abund.get_yaxis().tick_left() 
        ax_cyto_abund.grid(False)
        plt.xticks(fontsize=14, rotation=90)  
        plt.yticks(fontsize=14)
        plt.ylim(0, max([sum(x) for x in zip(cytokine_means, cytokine_std)]) *1.4)
        plt.ylabel(name + ' pg/mL', fontsize=14)
        plt.title('protein concentration', fontsize=12)
        plt.bar(y_pos, cytokine_means, align='center', alpha=0.75, yerr=cytokine_std, capsize=10, color=colors)
        plt.xticks(y_pos, empty,rotation=30)
        
# Call the function
        p_val_cyto_18hrs = protein_18hr_adj_pvals_dict[cytokine]
        p_val_cyto_48hrs = protein_48hr_adj_pvals_dict[cytokine]
        
        label_diff(0,1,p_val_message(p_val_cyto_18hrs),y_pos,cytokine_means, ax_cyto_abund, cytokine_std)
        label_diff(2,3,p_val_message(p_val_cyto_48hrs),y_pos,cytokine_means, ax_cyto_abund, cytokine_std)
        plt.tight_layout()


        # cytokine fold change
        ax2_cyto_fold = plt.subplot(2, 2, 2)
        ax2_cyto_fold.spines["top"].set_visible(False)  
        ax2_cyto_fold.spines["right"].set_visible(False) 
        y_pos = np.arange(len(conditions_2))
        ax2_cyto_fold.get_xaxis().tick_bottom()  
        ax2_cyto_fold.get_yaxis().tick_left() 
        ax2_cyto_fold.grid(False)
        plt.xticks(fontsize=14, rotation=90)  
        plt.yticks(fontsize=14)
        plt.ylabel(name + ' log2 fold-change', fontsize=14)
        # plt.xlabel("condition", fontsize=14)
        plt.title('protein virus/control', fontsize=12)
        plt.bar(y_pos, cytokine_fold_change, align='center', alpha=0.5, capsize=10, color='k')
        plt.xticks(y_pos, empty,rotation=30)
        plt.axhline(y=0.0, color='k', linestyle='-')
        plt.tight_layout()


        ax3_trans_abund = plt.subplot(2, 2, 3)
        ax3_trans_abund.spines["top"].set_visible(False)  
        ax3_trans_abund.spines["right"].set_visible(False) 
        y_pos = np.arange(len(conditions_1))
        ax3_trans_abund.get_xaxis().tick_bottom()  
        ax3_trans_abund.get_yaxis().tick_left() 
        ax3_trans_abund.grid(False)
        # plt.ylim([0.7,1.0])
        plt.xticks(fontsize=14, rotation=90)  
        plt.yticks(fontsize=14)
        plt.ylabel(gene_name + ' transcript counts', fontsize=14)
        # plt.xlabel("condition", fontsize=14)
        plt.title('transcript count', fontsize=12)
        plt.bar(y_pos, RNA_means, align='center', alpha=0.75, yerr=RNA_std, capsize=10, color=colors)
        plt.ylim(0, max([sum(x) for x in zip(RNA_means, RNA_std)]) *1.4)
        plt.xticks(y_pos, conditions_1,rotation=30)
        p_val_18_adj = DE_chip_c18_v18_df.loc[gene_name]['padj']
        p_val_48_adj = DE_chip_c48_v48_df.loc[gene_name]['padj']
#         label_diff(0,1,p_val_message(p_val_cyto_18hrs),[y_pos[0], y_pos[1]],cytokine_means, ax_cyto_abund, cytokine_std)
        label_diff(0,1,p_val_message(p_val_18_adj),y_pos,RNA_means, ax3_trans_abund, RNA_std)
        label_diff(2,3,p_val_message(p_val_48_adj),y_pos,RNA_means, ax3_trans_abund, RNA_std)
        plt.tight_layout()
        
        
        ax4_trans_fold = plt.subplot(2, 2, 4)
        ax4_trans_fold.spines["top"].set_visible(False)  
        ax4_trans_fold.spines["right"].set_visible(False) 
        y_pos = np.arange(len(conditions_2))
        ax4_trans_fold.get_xaxis().tick_bottom()  
        ax4_trans_fold.get_yaxis().tick_left() 
        ax4_trans_fold.grid(False)
        # plt.ylim([0.7,1.0])
        plt.xticks(fontsize=14, rotation=90)  
        plt.yticks(fontsize=14)
        plt.ylabel(gene_name + ' log2 fold-change', fontsize=14)
        # plt.xlabel("condition", fontsize=14)
        plt.title('transcript virus/control', fontsize=12)
        plt.bar(y_pos, RNA_fold_changes, align='center', alpha=0.5, capsize=10, color='k')
        plt.axhline(y=0.0, color='k', linestyle='-')



        plt.xticks(y_pos, conditions_2,rotation=30)
        plt.tight_layout()
        plt.savefig('../../../figs/cytokine_data_47-plex/' + name+'.pdf')
        plt.savefig('../../../figs/cytokine_data_47-plex/' + name+'.svg')
        plt.savefig('../../../figs/cytokine_data_47-plex/' + name+'.png')


        plt.show()
    except KeyError: # if gene not present in differential expression matrix (because it was filtered out)
        
        cytokine_control_18 = cytokine_data_df.loc[:,cytokine][1:4]
        cytokine_virus_18 = cytokine_data_df.loc[:,cytokine][4:7]
        cytokine_control_48 = cytokine_data_df.loc[:,cytokine][7:10]
        cytokine_virus_48 = cytokine_data_df.loc[:,cytokine][10:13]

        cytokine_means = [np.mean(cytokine_control_18), np.mean(cytokine_virus_18), np.mean(cytokine_control_48), np.mean(cytokine_virus_48)]
        cytokine_std = [np.std(cytokine_control_18), np.std(cytokine_virus_18), np.std(cytokine_control_48), np.std(cytokine_virus_48)]

        cytokine_fold_change = [log2_division(cytokine_means[1], cytokine_means[0]),log2_division(cytokine_means[3], cytokine_means[2])]

#         prot_idx = cytokine_data_df.columns.get_loc(cytokine)
#         gene_name = cytokine_data_df.loc['gene_name',:][prot_idx]

        control_18_names = [samp for samp in samples_control18_virus_18 if 'C' in samp]
        control_48_names = [samp for samp in samples_control48_virus_48 if 'C' in samp]

        virus_18_names = [samp for samp in samples_control18_virus_18 if 'V' in samp]
        virus_48_names = [samp for samp in samples_control48_virus_48 if 'V' in samp]

        figure(figsize=(8, 8), dpi = 300)  
        ax_cyto_abund = plt.subplot(2, 2, 1)
        ax_cyto_abund.spines["top"].set_visible(False)  
        ax_cyto_abund.spines["right"].set_visible(False) 
        y_pos = np.arange(len(conditions_1))
        ax_cyto_abund.get_xaxis().tick_bottom()  
        ax_cyto_abund.get_yaxis().tick_left() 
        ax_cyto_abund.grid(False)
        # plt.ylim([0.7,1.0])
        plt.xticks(fontsize=14, rotation=90)  
        plt.yticks(fontsize=14)
        plt.ylim(0, max([sum(x) for x in zip(cytokine_means, cytokine_std)]) *1.4)
        plt.ylabel(name + ' pg/mL', fontsize=14)
        # plt.xlabel("condition", fontsize=14)
        plt.title('protein concentration', fontsize=12)
        plt.bar(y_pos, cytokine_means, align='center', alpha=0.75, yerr=cytokine_std, capsize=10, color=colors)
        plt.xticks(y_pos, empty,rotation=30)
        
# Call the function
        p_val_cyto_18hrs = protein_18hr_adj_pvals_dict[cytokine]
        p_val_cyto_48hrs = protein_48hr_adj_pvals_dict[cytokine]
        
        label_diff(0,1,p_val_message(p_val_cyto_18hrs),y_pos,cytokine_means, ax_cyto_abund, cytokine_std)
        label_diff(2,3,p_val_message(p_val_cyto_48hrs),y_pos,cytokine_means, ax_cyto_abund, cytokine_std)

        plt.tight_layout()
        
        ax2_cyto_fold = plt.subplot(2, 2, 2)
        ax2_cyto_fold.spines["top"].set_visible(False)  
        ax2_cyto_fold.spines["right"].set_visible(False) 
        y_pos = np.arange(len(conditions_2))
        ax2_cyto_fold.get_xaxis().tick_bottom()  
        ax2_cyto_fold.get_yaxis().tick_left() 
        ax2_cyto_fold.grid(False)
        # plt.ylim([0.7,1.0])
        plt.xticks(fontsize=14, rotation=90)  
        plt.yticks(fontsize=14)
        plt.ylabel(name + ' log2 fold-change', fontsize=14)
        # plt.xlabel("condition", fontsize=14)
        plt.title('protein virus/control', fontsize=12)
        plt.bar(y_pos, cytokine_fold_change, align='center', alpha=0.5, capsize=10, color='k')
        plt.xticks(y_pos, empty,rotation=30)

        plt.axhline(y=0.0, color='k', linestyle='-')
        plt.tight_layout()
        # plt.savefig(output_dir+"energy_charge_mean.pdf")
        # plt.savefig(output_dir+"energy_charge_mean.svg")
        plt.savefig('../../../figs/cytokine_data_47-plex/' + name.replace(' ','')+'.pdf')
        plt.savefig('../../../figs/cytokine_data_47-plex/' + name.replace(' ','')+'.svg')
        plt.savefig('../../../figs/cytokine_data_47-plex/' + name.replace(' ','')+'.png')
        print("error" + cytokine)
        genes_filtered_out.append(cytokine)

```
